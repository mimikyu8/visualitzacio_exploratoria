Resumeix gràficament el data set DelayedFlights.csv

```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("DelayedFlightsNet.csv", index_col = 0)
```

```python
df.info()
```

Crea almenys una visualització per:

* Una variable categòrica (UniqueCarrier)

```python
sns.countplot( x = "UniqueCarrier", data = df, color = "BlueViolet").set(
    title = "Number of flights per unique carrier")

plt.savefig("flights_carrier.png")
```

* Una variable numèrica (ArrDelay)

```python
sns.boxplot(x = "ArrDelay", data = df, color = "DarkOrchid", fliersize = 0.5).set(
    title = "Arrival delay in minutes")

plt.savefig("arrdelay.png")
```

```python
sns.histplot(df, x = "DepDelay", color = "DarkMagenta", log_scale = True).set(
    title = "Departure delay in minutes. Logarithmic scale")

plt.savefig("depdelay.png")
```

* Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

```python
sns.violinplot(x = "ArrDelay", y = "UniqueCarrier", data = df, scale = "count").set(
    title = "Arrival Delay per Carrier")

plt.savefig("arrdelay_carrier.png")
```

* Dues variables numèriques (ArrDelay i DepDelay)

```python
sns.relplot(x = "ArrDelay", y = "DepDelay", data = df, alpha = 0.5).set(
    title = "Correlation of departure delay and arrival delay")

plt.savefig("depdelay_arrdelay.png")
```

```python
plot = sns.jointplot(x = "ArrDelay", y = "DepDelay", data = df, cmap = "Purples", kind = "hex",
                     xlim = (-15, 150), ylim = (-15, 150))
plot.fig.suptitle("Detail of correlation of departure delay and arrival delay")
plot.fig.subplots_adjust(top = 0.93)
plt.show()

plt.savefig("detail_depdelay_arrdelay.png")
```

* Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
sns.scatterplot(data = df, x = "ArrDelay", y = "DepDelay", size = "UniqueCarrier", alpha = 0.5,
                s = 20, sizes = (50, 500), hue = "UniqueCarrier"
               ).set(title = "Departure delay and arrival delay per Carrier")
plt.legend(bbox_to_anchor = (1.02, 1), loc = "upper left", borderaxespad = 0)
plt.show()

plt.savefig("depdelay_arrdelay_carrier.png")
```

```python
plot = sns.pairplot(df[["DepDelay", "ArrDelay", "AirTime"]])
plot.fig.suptitle("Departure delay, Arrival Delay and Air Time comparison")
plot.fig.subplots_adjust(top = 0.93)
plt.savefig("depdelay_arrdelay_airtime.png")
```

```python
plot = sns.relplot(data = df, x = "ArrDelay", y = "DepDelay", size = "UniqueCarrier", alpha = 0.5,
                s = 20, sizes = (50, 500), hue = "UniqueCarrier", col = "UniqueCarrier", col_wrap = 4)
               
plt.legend(bbox_to_anchor = (1.02, 1), loc = "upper left", borderaxespad = 0)
plot.fig.suptitle("Departure delay and arrival delay per Carrier", fontsize = 50)
plot.fig.subplots_adjust(top = 0.93)
plt.show()

plt.savefig("depdelay_arrdelay_carrier_sep.png")
```

* Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

```python
sns.scatterplot(data = df, x = "ArrDelay", y = "DepDelay", hue = "UniqueCarrier", size = "AirTime", 
                alpha = 0.5, s = 20, sizes = (50, 500)
               ).set(title = "Departure delay, arrival delay and airtime per Carrier")
plt.legend(bbox_to_anchor = (1.02, 1), loc = "upper left", borderaxespad = 0)
plt.show()

plt.savefig("depdelay_arrdelay_carrier_airtime.png")
```

```python
plot = sns.relplot(data = df, x = "ArrDelay", y = "DepDelay", hue = "UniqueCarrier", size = "AirTime",
            col = "UniqueCarrier", col_wrap = 4, alpha = 0.5, s = 20, sizes = (50, 500)
           )
plot.fig.suptitle("Departure delay, arrival delay and airtime per Carrier", fontsize = 50)
plt.legend(bbox_to_anchor = (1.02, 1), loc = "upper left", borderaxespad = 0, fontsize = 40)
plot.fig.subplots_adjust(top = 0.93)
plt.show()

plt.savefig("depdelay_arrdelay_carrier_airtime_sep.png")
```
